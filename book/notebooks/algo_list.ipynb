```python
# Read list of table names (use a dict to keep organized)
h = {}
for i in alg_list:
    s = i.split(' | ')
    h[s[1]] = {'Type': (s[2]), 'Geometry Requirement': (s[3]), 'Threshold Requirement': (s[4]), 'Temporal Requirement': (s[5]), 'Region': (s[6]), 'DOI/Reference': (s[7])}

# Stick Developer to Type so we can create a v basic ds to get things going
names = []
type_1 = []
for i in h:
    names.append(i)
    type_1.append(h[i]['Type'])

# Create v basic ds
d = {"Developer": {"dims": ("Developer"), "data": names}}
ds = xr.Dataset.from_dict(d)

# Assign the rest of the info to the ds
dd=[]
for i in ds.Developer.values:
    dd.append(ds.sel(Developer=i).assign(h[i]))
    
ds = xr.concat(dd, dim='Developer')
```
